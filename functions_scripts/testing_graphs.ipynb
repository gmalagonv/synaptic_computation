```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python


# Create an empty graph
G = nx.Graph()

# Add nodes
G.add_nodes_from([1, 2, 3, 4])

# Add edges
G.add_edges_from([(1, 2), (2, 3), (3, 4), (4, 1)])

# Draw the graph
nx.draw(G, with_labels=True, node_color='skyblue', node_size=800, font_size=12, font_color='black', font_weight='bold')
plt.show()
```

```python


# Create an empty graph
G = nx.DiGraph()

# Add nodes
G.add_nodes_from([1, 2, 3, 4])

# Add edges
G.add_edges_from([(1, 2), (2, 1), (2, 3), (3, 4), (4, 1)])

# Draw the graph
nx.draw(G, with_labels=True, node_color='skyblue', node_size=800, font_size=12, font_color='black', font_weight='bold')
plt.show()
```
